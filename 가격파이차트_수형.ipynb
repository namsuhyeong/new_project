```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정
plt.rcParams['font.family'] = 'NanumGothic'

# 폰트가 정상적으로 설정되었는지 확인
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
fontprop = fm.FontProperties(fname=font_path)

# 데이터 로드
file_path = './스포츠 및 엔터테인먼트03.csv'
data = pd.read_csv(file_path)

# 데이터 컬럼 확인
data.head(), data.columns
```

```python
data = data.dropna(subset=['댓글', '별점'])
```

```python
# 데이터 전처리: '가격' 열에서 숫자만 추출
import re
def extract_price(price_str):
    match = re.search(r'\d+', price_str)
    return int(match.group()) if match else None

data['가격'] = data['가격'].apply(extract_price).dropna().astype(int)
```

```python
# 가격 데이터 정규화
normalized_prices = (data['가격'] - data['가격'].min()) / (data['가격'].max() - data['가격'].min())
```

```python
# 정규화된 데이터를 범위에 따라 그룹화 (실제 가격 범위 레이블 사용)
def categorize_normalized_price(price):
    if price <= (5000 - data['가격'].min()) / (data['가격'].max() - data['가격'].min()):
        return '0~5000원'
    elif price <= (10000 - data['가격'].min()) / (data['가격'].max() - data['가격'].min()):
        return '5001~10000원'
    elif price <= (20000 - data['가격'].min()) / (data['가격'].max() - data['가격'].min()):
        return '10001~20000원'
    elif price <= (30000 - data['가격'].min()) / (data['가격'].max() - data['가격'].min()):
        return '20001~30000원'
    else:
        return '그외'

# 새로운 컬럼 생성
data['정규화된 가격 범위'] = normalized_prices.apply(categorize_normalized_price)
```

```python
# 각 그룹의 비율 계산
price_range_counts = data['정규화된 가격 범위'].value_counts()
price_range_ratios = price_range_counts / price_range_counts.sum()

# 한글 폰트 설정 (NanumGothic)
plt.rcParams['font.family'] = 'NanumGothic'
```

```python
# 파이차트 생성
plt.figure(figsize=(10, 8))
plt.pie(price_range_ratios, labels=price_range_ratios.index, autopct='%1.1f%%', startangle=140)
plt.title('정규화된 가격 범위별 비율 파이차트')
plt.show()
```
